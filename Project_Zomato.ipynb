## Zomato Dataset Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the Data With the Help of Pandas

df = pd.read_csv('zomato.csv', encoding = 'latin-1')
```

```python
df.head()
```

```python
# These are my int Features and their descriptive statisics
df.describe()
```

```python
# To check my Columns and the Datatypes of each Feature
df.info()
```

```python
#To Check the Null Value or Missing Value in the Dataset
df.isnull().sum()
```

```python
# To check the null values with the help of heatmap.
sns.heatmap(df.isnull() , yticklabels = False , cbar = False , cmap ='coolwarm')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_country.shape
```

```python
#To check The datatypes of the the Country
df_country.info()
```

```python
# To Check the null or Missing Values
df_country.isnull().sum()
```

There is No null value in the dataset

In Both Dataset i have the country code so i Will Merge Both Of Them. ( df & df_country )

```python
# This is my final record by merging both of the datasets
Final_df = pd.merge(df , df_country , how = 'outer' , on ='Country Code')
```

```python
Final_df.head()
```

```python
Final_df.columns
```

```python
Country_No = Final_df['Country'].value_counts().values
```

```python
Country_Name = Final_df['Country'].value_counts().index 
```

```python
# Top Countries Using Zomato
plt.pie(Country_No[:3] , labels = Country_Name[:3]  , autopct = '%1.2f%%'  )
```

Observations - It is Observed that Approximately 95% users of Zomato are from india Followed by United States and United Kingdom

```python
Ratings = Final_df.groupby(['Aggregate rating' , 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Total_Restaurant'})
```

```python
Ratings
```

Observations - 
2148 No of Peoples didnot Rated the Zomato

When Rating is Between (1.8 - 2.4) ---------->Poor

When Rating is Between (2.5 - 3.4) ---------->Average

When Rating is Between (3.5 - 3.9) ---------->Good

When Rating is Between (4.0 - 4.4) ---------->Very Good

When Rating is Between (4.4 - 4.9) ---------->Excellent






```python
sns.barplot(x = 'Aggregate rating' , y = 'Total_Restaurant' , data = Ratings , hue = 'Rating color' , palette = ['White' , 'Red' , 'Orange' , 'Yellow' , 'Green' , 'Green'])
```

Observations =

1 - Count of not rated is very high

2 - Maximum No of peoples are rated between 2.5 to 3.5

```python
sns.countplot(x = 'Rating color' , hue = 'Rating color' , palette = ['Blue' , 'Red' , 'Orange' , 'Yellow' , 'Green' , 'Green'] , data = Ratings)
```

Observations - Orange Rating has Maximum no of frequency in the Data which States that the Maximum number of peoples giving between [2.5 to 3.5]
